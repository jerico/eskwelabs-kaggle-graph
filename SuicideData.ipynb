```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('master.csv')
df
```

```python
sns.lineplot(df['year'], df['suicides_no'])
#Year-by-year trend of suicide 
```

```python
plt.figure(figsize=(12,6))
df.groupby('country')['suicides_no'].sum().sort_values(ascending=False).plot.bar()
#Total Number of Suicides Per Country Since Year xx
```

```python
#Total Number of Suicides Per Generation Since Year xx

#Pyplot Version
#Sorted descending
#plt.figure(figsize=(12,6))
#df.groupby('generation')['suicides_no'].sum().sort_values(ascending=False).plot.bar()

#Seaborn Version
#Sorted according to generation
sns.barplot(data=df, x="generation", y="suicides_no", order=["G.I. Generation", "Silent", "Boomers", "Generation X", "Millenials", "Generation Z"])
```

```python
#Total Number of Suicides Per Age Group Since Year xx

#Pyplot Version
#Sorted descending
#plt.figure(figsize=(12,6))
#df.groupby('age')['suicides_no'].sum().sort_values(ascending=False).plot.bar()

#Seaborn Version
#Sorted according to age group
sns.barplot(data=df, x="age", y="suicides_no", order=["5-14 years","15-24 years","25-34 years", "35-54 years", "55-74 years", "75+ years"])
```

```python
#Total Number of Suicides Per Sex Since Year xx

#Pyplot Version
#plt.figure(figsize=(12,6))
#df.groupby('sex')['suicides_no'].sum().sort_values(ascending=False).plot.bar()

#Seaborn Version
sns.barplot(data=df, x="sex", y="suicides_no")
```
